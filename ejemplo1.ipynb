## ** Cambiar de BGR a RGB**

```
img_RGB=cv2.cvtColor(img_BGR,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,10))
plt.imshow(img_RGB)
plt.xticks([])
plt.yticks([])
plt.show()
```

## **Convertir a escala de grices **

```
img_GRAY=cv2.cvtColor(img_RGB,cv2.COLOR_RGB2GRAY)

plt.figure(figsize=(10,10))
plt.imshow(img_GRAY)
plt.xticks([])
plt.yticks([])
plt.show()
```

```
img_GRAY=cv2.cvtColor(img_RGB,cv2.COLOR_RGB2GRAY)

plt.figure(figsize=(10,10))
plt.imshow(img_GRAY,cmap='gray')
plt.xticks([])
plt.yticks([])
plt.show()
```

### **Crear y usar metodos para mostrar las imagenes **

```
def display(a, b, title1 = "Original", title2= "Editado"):
    plt.figure(figsize=(14,14))
    plt.subplot(121), plt.imshow(a), plt.title(title1)
    plt.xticks([]), plt.yticks([])
    plt.subplot(122), plt.imshow(b), plt.title(title2)
    plt.xticks([]), plt.yticks([])
    plt.show()
    
    
    
def display_one(a, title1= "Original"):
    plt.imshow(a), plt.title(title1)
    plt.show()
    

display_one(img_BGR)



display(img_BGR, img_GRAY, "BGR", "GRAY")
```

```
def display4(a, b, c, d, title1 = "Original", title2= "Editado", title3= "Editado", title4= "Editado"):
    plt.figure(figsize=(14,14))
    plt.subplot(221), plt.imshow(a), plt.title(title1)
    plt.xticks([]), plt.yticks([])
    plt.subplot(222), plt.imshow(b), plt.title(title2)
    plt.xticks([]), plt.yticks([])
    plt.subplot(223), plt.imshow(c), plt.title(title3)
    plt.xticks([]), plt.yticks([])
    plt.subplot(224), plt.imshow(d, cmap="gray"), plt.title(title4)
    plt.xticks([]), plt.yticks([])
    plt.show()
    
display4(img_BGR,img_RGB,img_GRAY,img_GRAY, "BGR", "RGB", "GRAY", "GRAY2")
```

```
def estadisticas(img):
  
  img_size=img.shape
  
  if len(img_size)>2:
    print("image chanels", img_size[2])
    
  else:
    print("image chanels", 1)
    
  print("image W",img_size[1])
  print("image H",img_size[0])
  print("min value",img.min())
  print("max value",img.max())
  print("prom",img.mean())
  print("type",img.dtype)
  
estadisticas(img_BGR)
estadisticas(img_GRAY)
```
